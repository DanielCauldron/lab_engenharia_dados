```python
# Explorando dados de clientes - Notebook inicial

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configurações visuais
sns.set(style="whitegrid")

# Carregar dados transformados
path = Path("../data/processed/clientes_transformado.csv")
df = pd.read_csv(path)

# Mostrar as primeiras linhas
print("Primeiras 5 linhas:")
display(df.head())

# Informações gerais do DataFrame
print("\nInformações gerais:")
print(df.info())

# Estatísticas básicas para colunas numéricas
print("\nEstatísticas descritivas:")
display(df.describe())

# Contagem de clientes por estado (se existir coluna 'state', senão ignore)
if "state" in df.columns:
    plt.figure(figsize=(12, 5))
    sns.countplot(data=df, x="state", order=df["state"].value_counts().index)
    plt.title("Quantidade de clientes por estado")
    plt.xticks(rotation=45)
    plt.show()

# Distribuição dos nomes - quantos clientes por first_name
if "first_name" in df.columns:
    nomes_mais_comuns = df["first_name"].value_counts().head(10)
    plt.figure(figsize=(10, 4))
    sns.barplot(x=nomes_mais_comuns.values, y=nomes_mais_comuns.index)
    plt.title("Top 10 primeiros nomes mais comuns")
    plt.show()

# Emails únicos
print(f"\nTotal de emails únicos: {df['email'].nunique()}")

# Exemplo: clientes com telefone vazio
telefones_vazios = df["telefone"].isna().sum()
print(f"\nClientes sem telefone informado: {telefones_vazios}")

# Mostrar exemplos aleatórios
print("\nExemplos aleatórios de clientes:")
display(df.sample(5))
```
